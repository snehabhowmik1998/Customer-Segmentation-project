```python
from sklearn.preprocessing import LabelEncoder
```

```python
lr = LabelEncoder()
```

```python
df["city_numbers"] = lr.fit_transform(df["City"])
```

```python
df["Most Purchased Product_numbers"] = lr.fit_transform(df["Most Purchased Product"])
```

```python
df
```

```python
cleaned_data = df.drop(["City","Most Purchased Product"], axis =1)
```

```python
cleaned_data
```

```python
cleaned_data.info()
```

```python
from sklearn.cluster import KMeans
```

```python
x= cleaned_data[["Annual Revenue (k$)", "Spending Score (1-100)"]]
```

```python
inertia = []
for i in range(1,20):
    km = KMeans(n_clusters=i, init='k-means++' )
    km.fit(x)
    inertia.append(km.inertia_)
```

```python
inertia
```

```python
plt.plot(range(1,20),inertia,"bo-")
plt.title("Graphical representation of elbow method")
plt.xlabel("No. of clusters")
plt.ylabel("Inertia's")
plt.show()
```

## In the above graph from the elbow methoad we came to the conclusion that best no.of clusters is 5

```python
km_model = KMeans(n_clusters=5,init='k-means++',verbose=0)
```

```python
km_model.fit(x)
```

```python
pred = km_model.predict(x)
```

```python
pred
```

```python
len(pred)
```

```python
label = km.labels_
label
```

```python
centroids=km_model.cluster_centers_
centroids
```
